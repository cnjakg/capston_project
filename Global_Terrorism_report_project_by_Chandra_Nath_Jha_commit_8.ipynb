##**Q.4- Are there any patterns or clusters in the data that suggest specific regions or periods of high terrorist activity?**

```python
df_region_year=df2.groupby(['year','region']).agg({'total_death':'sum','wounded':'sum'}).sort_values(by='total_death',ascending=False).head(100)
df_region_year.head(10)
```

```python
# Reset index for easier plotting
df_region_year = df_region_year.reset_index()

# Pivot the data to get regions as columns
pivot_data = df_region_year.pivot(index='year', columns='region', values='total_death')

# Plot the stacked bar chart
plt.figure(figsize=(12, 4))
pivot_data.plot(kind='bar', stacked=True, colormap='viridis', ax=plt.gca())

plt.title('Total Deaths by Region Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Deaths')
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
